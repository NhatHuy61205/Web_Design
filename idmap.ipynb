```python
import asyncio
from requests_html import AsyncHTMLSession
from bs4 import BeautifulSoup

async def fetch_html():
    session = AsyncHTMLSession()
    try:
        response = await session.get("https://www.vietnamairlines.com/vn/vi/destinations/viet-nam")
        await response.html.arender()  # Xử lý JavaScript
        html = response.html.html
        return html
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        await session.close()


html = await fetch_html()

soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
svg_element = soup.select_one('body > form > div:nth-of-type(11) > div > div:nth-of-type(2) > div > div > div:nth-of-type(1) > svg')


if svg_element:
    print(svg_element.prettify())
else:
    print("SVG element not found")
```

```python
import csv
with open('idmap.csv', 'w', newline='') as csvfile:
    fieldnames = ['id', 'path_d']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
   
    writer.writeheader()

    if svg_element:
        g_elements = svg_element.find_all('g')
        for g in g_elements:
            g_id = g.get('id')
            if g_id:
                paths = g.find_all('path')
                for path in paths:
                    path_d = path.get('d')
                    if path_d:
                        writer.writerow({'id': g_id, 'path_d': path_d})
    else:
        print("SVG element not found")
```
